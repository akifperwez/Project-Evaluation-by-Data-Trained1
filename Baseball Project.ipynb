# Data Cleansing

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
#Removing outliers using IQR Method
df_new=df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]
```

```python
df_new.shape
```

```python
loss_percentage=(30-20)/30*100
print(loss_percentage)
```

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
thresold=3
print(np.where(z>3))
```

```python
df_new1=df[(z<3).all(axis=1)]
```

```python
df_new1.shape
```

```python
loss_percentage=(30-29)/30*100
print(loss_percentage)
```

```python
df_new1.skew()
```

```python
df_new1["H"]=np.log(df_new1["H"])
df_new1["SV"]=np.log(df_new1["SV"])
df_new1["E"]=np.log(df_new1["E"])
```

```python
df_new1.skew()
```

```python
x=df_new1.drop(["W"],axis=1)
y=df_new1["W"]
```

# Finding The Best Random State

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
maxAucc=0
maxRS=0
for i in range(1,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.23,random_state=i)
    le=LinearRegression()
    le.fit(x_train,y_train)
    pred=le.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAucc:
        maxAucc=acc
        maxRS=i
print("Best Accuracy is ",maxAucc,"on random_state",maxRS)
```

# Data Preprocessing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.23,random_state=30)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

x_trains = ss.fit_transform(x_train)
x_tests = ss.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_trains,y_train)
pred=lr.predict(x_tests)
print(r2_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_trains,y_train)
pred=rf.predict(x_tests)
print(r2_score(y_test,pred))
```

```python
from sklearn.ensemble import AdaBoostRegressor
ad=AdaBoostRegressor()
ad.fit(x_trains,y_train)
pred=ad.predict(x_tests)
print(r2_score(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
kn.fit(x_trains,y_train)
pred=kn.predict(x_tests)
print(r2_score(y_test,pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scr=cross_val_score(lr,x,y,cv=5)
print("Cross validation Score of Linear Regression",scr.mean())
```

```python
scr=cross_val_score(rf,x,y,cv=5)
print("Cross validation Score of Random Forest Regressor",scr.mean())
```

```python
scr=cross_val_score(ad,x,y,cv=5)
print("Cross validation Score of AdaBoost Regressor",scr.mean())
```

```python
scr=cross_val_score(kn,x,y,cv=5)
print("Cross validation Score of Kneighbors Regressor",scr.mean())
```

We do Hyper Parameter Tuning with Random Forest Regressor because it has the least difference

```python
GCV=GridSearchCV(LinearRegression(),parameters,cv=5) #tuning 
```

```python
GCV.best_params_
```

```python
lr_mod=LinearRegression(normalize="bool",n_jobs=1)

lr_mod.fit(x_trains,y_train)
pred=lr_mod.predict(x_tests)
print(r2_score(y_test,pred)*100)
```
