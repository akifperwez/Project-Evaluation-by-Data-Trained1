# Finding the best Random State

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
maxAucc=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    le=LogisticRegression()
    le.fit(x_train,y_train)
    pred=le.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAucc:
        maxAucc=acc
        maxRS=i
print("Best Accuracy is ",maxAucc,"on random_state",maxRS)
```

# Creating Train Test Split

```python
x_train,y_train,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=110)
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
print(accuracy_score(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
pred=dtc.predict(x_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
pred=svc.predict(x_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
kn.fit(x_train,y_train)
pred=kn.predict(x_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scr=cross_val_score(lr,x,y,cv=5)
print("Cross validation Score of logistic Regression",scr.mean()),3
```

```python
scr=cross_val_score(dtc,x,y,cv=5)
print("Cross validation Score of Decision Tree Classifier",scr.mean())
```

```python
scr=cross_val_score(rf,x,y,cv=5)
print("Cross validation Score of Random Forest Classifier",scr.mean())
```

```python
scr=cross_val_score(svc,x,y,cv=5)
print("Cross validation Score of Support Vector Classifier",scr.mean())
```

```python
scr=cross_val_score(kn,x,y,cv=5)
print("Cross validation Score of KNeighbors Classifier",scr.mean())
```

Here,Minimum difference in accuracy and cross validation is from Logistic, Kneighbors, we will do hyper tuning with both of them

# Hyper Parameter-Tuning

```python
from sklearn.model_selection import GridSearchCV
```

# Logistic Regression

```python
#creating parameter list to pass in GridSearchCV

parameters={"C":np.arange(2,15),
           "penalty":["l1", "l2","elasticnet"]}
```

```python
GCV=GridSearchCV(LogisticRegression(),parameters,cv=5)
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python
lr_mod=LogisticRegression(penalty="l2",C=8,random_state=110)

lr_mod.fit(x_train,y_train)
pred=lr_mod.predict(x_test)
print(accuracy_score(y_test,pred)*100)
```

# KNeighbors Classifier

```python
parameters={"n_neighbors":np.arange(2,15),
           "algorithm":["auto", "ball_tree","kd_tree","brute"]}
```

```python
GCV=GridSearchCV(KNeighborsClassifier(),parameters,cv=5)
GCV.fit(x_train,y_train)
GCV.best_params_
```

```python
kn_mod=KNeighborsClassifier(algorithm="auto",n_neighbors=10)

kn_mod.fit(x_train,y_train)
pred=kn_mod.predict(x_test)
print(accuracy_score(y_test,pred)*100)
```

# Plotting AUC and ROC Curve

```python
from sklearn.metrics import plot_roc_curve
```

```python
plot_roc_curve(lr_mod,x_test,y_test)
plt.title("Receiver Operating Characterstics")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

```python
plot_roc_curve(kn_mod,x_test,y_test)
plt.title("Receiver Operating Characterstics")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```
